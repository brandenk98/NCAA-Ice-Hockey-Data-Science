```python
### Getting the CSV file for the  season 19-20
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
season1920 = pd.read_csv('data/Season19-20.csv', sep=',', usecols=['players','year', 'school','position', 'goals','assists','points','powerplay_points','penalties','blocks', 'minus','plus'])
season1920
```

```python
NHLsigned1920 = pd.read_csv('data/NHLsigned20192020.csv', sep=',', usecols=['players','year','school','position', 'goals','assists','points','powerplay_points','penalties','blocks','plus','minus'])
NHLsigned1920
```

```python
### creating a scatterplot of players to signed players



#### THIS SCATTER PLOT ONLY LOOKS AT ASSISTS TO POINTS
```

```python
ax = season1920.reset_index().plot(kind='scatter', x='assists', y='points',
                                           color='Red', label='All players', s=50)

NHLsigned1920.reset_index().plot(kind='scatter', x='assists', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
### creating a scatterplot of players to signed players

### THIS SCATTER PLOT ONLY  LOOKS AT GOALS TO PLAYERS
```

```python
ax = season1920.reset_index().plot(kind='scatter', x='goals', y='points',
                                           color='Red', label='All players', s=50)

NHLsigned1920.reset_index().plot(kind='scatter', x='goals', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
### THESE SCATTER PLOTS SHOW NO CLUSTERS, BUT THIS IS JUST ANALYSIS ON ALL POSITIONS


    ###NOW WE  LOOK AT FORWARDS
```

```python
###Sorting to only forwards

forwards = season1920[season1920['position']=='F']
signedFowards = NHLsigned1920[NHLsigned1920['position']=='F']
```

```python
###Scatter plot for  ASSISTS
ax = forwards.reset_index().plot(kind='scatter', x='assists', y='points',
                                           color='Red', label='All players', s=50)

signedFowards.reset_index().plot(kind='scatter', x='assists', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
##Scatter plot for GOALS

ax = forwards.reset_index().plot(kind='scatter', x='goals', y='points',
                                           color='Red', label='All players', s=50)

signedFowards.reset_index().plot(kind='scatter', x='goals', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```
