```python
import altair as alt
import pandas as pd
```

```python
signedData0 = pd.read_csv('data/NHLsigned20192020.csv', sep=',', usecols=['plus/minus','points_per_game'])
signedData1 = pd.read_csv('data/NHLsigned20182019.csv', sep=',', usecols=['plus/minus','points_per_game'])
signedData2 = pd.read_csv('data/NHLsigned20172018.csv', sep=',', usecols=['plus/minus','points_per_game'])
signedData3 = pd.read_csv('data/NHLsigned20162017.csv', sep=',', usecols=['plus/minus','points_per_game'])
signedData4 = pd.read_csv('data/NHLsigned20152016.csv', sep=',', usecols=['plus/minus','points_per_game'])
```

```python
###2019-2020
brush = alt.selection_interval()
signed_chart0 = alt.Chart(signedData0).mark_point().encode(
    x='points_per_game',
    y='plus/minus',
    color=alt.condition(brush, 'Origin:N', alt.value('red'))
    
).add_selection(brush)


####2018-2019
signed_chart1 = alt.Chart(signedData1).mark_point().encode(
    x='points_per_game',
    y='plus/minus',
    color=alt.condition(brush, 'Origin:N', alt.value('green'))
    
).add_selection(brush)

###2017-2018
signed_chart2 = alt.Chart(signedData2).mark_point().encode(
    x='points_per_game',
    y='plus/minus',
    color=alt.condition(brush, 'Origin:N', alt.value('blue'))
    
).add_selection(brush)

###2016-2017
signed_chart3 = alt.Chart(signedData3).mark_point().encode(
    x='points_per_game',
    y='plus/minus',
    color=alt.condition(brush, 'Origin:N', alt.value('yellow'))
    
).add_selection(brush)


###2015-2016
signed_chart4 = alt.Chart(signedData4).mark_point().encode(
    x='points_per_game:Q',
    y='plus/minus:Q',
    color=alt.condition(brush, 'Origin:O', alt.value('black'))
    
).add_selection(brush)



signed_chart0 & signed_chart1 & signed_chart2 & signed_chart3 & signed_chart4
```
