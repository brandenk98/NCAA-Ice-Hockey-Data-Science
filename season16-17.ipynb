```python
### Getting the CSV file for the  season 16-17
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
season1617 = pd.read_csv('data/Season16-17.csv', sep=',', usecols=['players', 'school','position', 'goals','assists','points','powerplay_points','penalties','blocks', 'minus','plus'])
season1617
```

```python
NHLsigned1617 = pd.read_csv('data/NHLsigned20162017.csv', sep=',', usecols=['players','school','position', 'goals','assists','points','powerplay_points','penalties','blocks','plus','minus'])
NHLsigned1617
```

```python
### creating a scatterplot of players to signed players



#### THIS SCATTER PLOT ONLY LOOKS AT ASSISTS TO POINTS
```

```python
ax = season1617.reset_index().plot(kind='scatter', x='assists', y='points',
                                           color='Red', label='All players', s=50)

NHLsigned1617.reset_index().plot(kind='scatter', x='assists', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
### creating a scatterplot of players to signed players

### THIS SCATTER PLOT ONLY  LOOKS AT GOALS TO PLAYERS
```

```python
ax = season1617.reset_index().plot(kind='scatter', x='goals', y='points',
                                           color='Red', label='All players', s=50)

NHLsigned1617.reset_index().plot(kind='scatter', x='goals', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
### THESE SCATTER PLOTS SHOW NO CLUSTERS, BUT THIS IS JUST ANALYSIS ON ALL POSITIONS


    ###NOW WE  LOOK AT FORWARDS
```

```python
###Sorting to only forwards

forwards = season1617[season1617['position']=='F']
signedFowards = NHLsigned1617[NHLsigned1617['position']=='F']
```

```python
###Scatter plot for  ASSISTS
ax = forwards.reset_index().plot(kind='scatter', x='assists', y='points',
                                           color='Red', label='All players', s=50)

signedFowards.reset_index().plot(kind='scatter', x='assists', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
##Scatter plot for GOALS

ax = forwards.reset_index().plot(kind='scatter', x='goals', y='points',
                                           color='Red', label='All players', s=50)

signedFowards.reset_index().plot(kind='scatter', x='goals', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
##Sorting only Defense
defense = season1617[season1617['position']=='D']
signedDefense = NHLsigned1617[NHLsigned1617['position']=='D']
```

```python
###Scatter plot for  ASSISTS
ax = defense.reset_index().plot(kind='scatter', x='assists', y='points',
                                           color='Red', label='All players', s=50)

signedDefense.reset_index().plot(kind='scatter', x='assists', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
##Scatter plot for GOALS

ax = defense.reset_index().plot(kind='scatter', x='goals', y='points',
                                           color='Red', label='All players', s=50)

signedDefense.reset_index().plot(kind='scatter', x='goals', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```
