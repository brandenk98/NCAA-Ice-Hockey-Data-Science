```python
### Getting the CSV file for the  season 15-16
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
season1516 = pd.read_csv('data/Season15-16.csv', sep=',', usecols=['players', 'school','position', 'goals','assists','points','powerplay_points','penalties','blocks', 'minus','plus'])
season1516
```

```python
NHLsigned1516 = pd.read_csv('data/NHLsigned20152016.csv', sep=',', usecols=['players','school','position', 'goals','assists','points','powerplay_points','penalties','blocks','plus','minus'])
NHLsigned1516
```

```python
### creating a scatterplot of players to signed players



#### THIS SCATTER PLOT ONLY LOOKS AT ASSISTS TO POINTS
```

```python
ax = season1516.reset_index().plot(kind='scatter', x='assists', y='points',
                                           color='Red', label='All players',title='2015-16', s=50)

NHLsigned1516.reset_index().plot(kind='scatter', x='assists', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
### creating a scatterplot of players to signed players

### THIS SCATTER PLOT ONLY  LOOKS AT GOALS TO PLAYERS
```

```python
ax = season1516.reset_index().plot(kind='scatter', x='goals', y='points',
                                           color='Red', label='All players', title='2015-16',s=50)

NHLsigned1516.reset_index().plot(kind='scatter', x='goals', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
### THESE SCATTER PLOTS SHOW NO CLUSTERS, BUT THIS IS JUST ANALYSIS ON ALL POSITIONS


    ###NOW WE  LOOK AT FORWARDS
```

```python
###Sorting to only forwards

forwards = season1516[season1516['position']=='F']
signedFowards = NHLsigned1516[NHLsigned1516['position']=='F']
```

```python
###Scatter plot for  ASSISTS
ax = forwards.reset_index().plot(kind='scatter', x='assists', y='points',
                                           color='Red', label='All players', s=50)

signedFowards.reset_index().plot(kind='scatter', x='assists', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
##Scatter plot for GOALS

ax = forwards.reset_index().plot(kind='scatter', x='goals', y='points',
                                           color='Red', label='All players', s=50)

signedFowards.reset_index().plot(kind='scatter', x='goals', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
##Sorting only Defense
defense = season1516[season1516['position']=='D']
signedDefense = NHLsigned1516[NHLsigned1516['position']=='D']
```

```python
###Scatter plot for  ASSISTS
ax = defense.reset_index().plot(kind='scatter', x='assists', y='points',
                                           color='Red', label='All players', s=50)

signedDefense.reset_index().plot(kind='scatter', x='assists', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```

```python
##Scatter plot for GOALS

ax = defense.reset_index().plot(kind='scatter', x='goals', y='points',
                                           color='Red', label='All players', s=50)

signedDefense.reset_index().plot(kind='scatter', x='goals', y='points',
                                          color='Blue', label='NHL signed', ax=ax, s=50)
```
